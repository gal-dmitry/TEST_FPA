#### Идея решения "в лоб"

Для прямого решения по предсказанию `ocean_proximity` мы могли бы использовать только лишь значения `longitude` и `latitude`. Из графика ниже видно, что можно построить границы кластеров, используя размеченные точки. С помощью границ можно было бы поделить плоскость на несколько областей и присваивать значение `ocean_proximity` в зависимости от того, в какую область попадает новая точка.

```python
def visualize_clusters(df):
    new_df = df.copy()
    labels, categories = factorize(new_df["ocean_proximity"])
    new_df["labels"] = labels
    cut_df = new_df[new_df["labels"] != -1]
    
    
    d = {"<1H OCEAN": 'purple', 
         "INLAND": 'green', 
         "ISLAND": 'red', 
         "NEAR BAY": 'yellow',
         "NEAR OCEAN": 'blue',
         None: 'black'}
    
    for key, value in d.items():
        print(f"{key}: {value}")
    
    new_df['colors'] = new_df['ocean_proximity'].map(d)
    cut_df['colors'] = cut_df['ocean_proximity'].map(d)   
    
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7.5))
    
    # left
    scatter1 = ax1.scatter(x = cut_df.longitude,
                             y = cut_df.latitude,
                             c = cut_df.colors,
                             alpha = 0.25)

    ax1.set_title("Только известные значения")
    ax1.set_xlabel("longitude")
    ax1.set_ylabel("latitude")
    
    # right
    scatter2 = ax2.scatter(x = new_df.longitude,
                             y = new_df.latitude,
                             c = new_df.colors,
                             alpha = 0.25)

    ax2.set_title("Весь датасет")
    ax2.set_xlabel("longitude")
    ax2.set_ylabel("latitude")
    
    plt.show()
```

- Недостаток нейросети в нашей задаче случае - очень маленький размер отдельных классов. Зато, кроме координат, мы можем добавить 2 остальных признака

##### Dataset preparation

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
def plt_matrix(confusion_matrix):

    fig, ax = plt.subplots()    
    ax.matshow(confusion_matrix, cmap=plt.cm.Blues)   
    h, w = confusion_matrix.shape
    for i in range(h):
        for j in range(w):
            c = confusion_matrix[j,i]
            ax.text(i, j, str(c), va='center', ha='center')
            

def separate_dataset(df): 
    
    labels, categories = factorize(df["ocean_proximity"])
    df["labels"] = labels
    pairs = df[["ocean_proximity", "labels"]].drop_duplicates()
    
    index = df.loc[pd.isna(df["ocean_proximity"]), :].index
    
    labeled_part = df.loc[~df.index.isin(index)]
    unlabeled_part = df.loc[index]  
    
    return labeled_part, unlabeled_part, pairs


def df2np(df, only_coo=True):
    
    if only_coo:
        X = np.array([[row.longitude, row.latitude] for _, row in df.iterrows()])
    else:
        X = np.array([[row.longitude, row.latitude, row.median_house_value, row.population] 
                      for _, row in df.iterrows()])
        
    y = np.array([row.labels for _, row in df.iterrows()])
    return X, y


def split_train_test(df, only_coo=True):
    X, y = df2np(df, only_coo=only_coo)    
    skf = StratifiedKFold(n_splits=2)
    for train_index, test_index in skf.split(X, y):  
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
    
    return X_train, X_test, y_train, y_test

            

class Dataset:
    def __init__(self, df):
        
        labeled_part, unlabeled_part, pairs = separate_dataset(df)
        
        X_train_coo, X_test_coo, y_train, y_test = split_train_test(labeled_part, only_coo=True)
        X_unlabeled_coo, _ = df2np(unlabeled_part, only_coo=True)
        X_labeled_coo, y = df2np(labeled_part, only_coo=True)
        
        X_train, X_test, _, _ = split_train_test(labeled_part, only_coo=False)
        X_unlabeled, _ = df2np(unlabeled_part, only_coo=False)
        X_labeled, _ = df2np(labeled_part, only_coo=False)
        
        self.pairs = pairs.set_index("labels").to_dict()['ocean_proximity']
        
        self.y = y
        self.y_train = y_train
        self.y_test = y_test
        
        self.X_train_coo = X_train_coo
        self.X_test_coo = X_test_coo
        
        self.X_train = X_train
        self.X_test = X_test
        
        self.X_unlabeled_coo = X_unlabeled_coo
        self.X_unlabeled = X_unlabeled
        
        self.X_labeled_coo = X_labeled_coo
        self.X_labeled = X_labeled
        
        # scale data
        scaler = StandardScaler()
        scaler.fit(X_train_coo)
        self.X_train_coo_scaled = scaler.transform(X_train_coo)
        self.X_test_coo_scaled = scaler.transform(X_test_coo)
        
        scaler = StandardScaler()
        scaler.fit(X_labeled_coo)
        self.X_labeled_coo_scaled = scaler.transform(X_labeled_coo)
        self.X_unlabeled_coo_scaled = scaler.transform(X_unlabeled_coo)
        
        scaler = StandardScaler()
        scaler.fit(X_train) 
        self.X_train_scaled = scaler.transform(X_train)
        self.X_test_scaled = scaler.transform(X_test)
        
        scaler = StandardScaler()
        scaler.fit(X_labeled)
        self.X_labeled_scaled = scaler.transform(X_labeled)
        self.X_unlabeled_scaled = scaler.transform(X_unlabeled)
```

##### Solver

```python
class Solver:
    
    def __init__(self, clf):
        self.clf = clf
        
    def train(self, X_train, y_train):
        self.clf.fit(X_train, y_train)
    
    def test(self, X_test, y_test):
        y_pred = self.clf.predict(X_test)
        print(classification_report(y_test, y_pred))
        c_matrix = confusion_matrix(y_test, y_pred)
        plt_matrix(c_matrix)
    
    def predict(self, X_labeled, y, X_unlabeled):
        self.clf.fit(X_labeled, y)
        y_pred = self.clf.predict(X_unlabeled)
        return y_pred
```

```python
knn = KNeighborsClassifier(n_neighbors=10, weights='distance')
KNN_Solver = Solver(knn)

KNN_Solver.train(dataset.X_train_coo, dataset.y_train)
KNN_Solver.test(dataset.X_test_coo, dataset.y_test)
```

### TASK 2

Исправить null-значения **ocean_proximity** в самой БД. Условие: нельзя заменять таблицу в БД на новую, только через обновление значений с помощью SQL;

```python
neighborhoods_filled_df = mlp_df.copy()
```

```python
def sql_fillna(neighborhoods_df, neighborhoods_filled_df):

    index = neighborhoods_df.loc[pd.isna(neighborhoods_df["ocean_proximity"]), :].index

    records_to_upd = []
    for _, row in neighborhoods_filled_df.loc[index].iterrows():
        records_to_upd.append((row.id, row.ocean_proximity))
        
    src = "test_project.db"
    dst = "test_project_filled.db"
    shutil.copyfile(src, dst)
    print(f"{dst} created")
    
    engine_v1 = create_engine(f'sqlite:///{dst}')
    
    for (_id, cluster_label) in tqdm(records_to_upd):
        with engine_v1.begin() as conn:
            sql = f"""UPDATE neighborhoods SET ocean_proximity = "{cluster_label}" WHERE id = {_id}"""
            conn.execute(sql)
            
```

### TASK 3

С помощью SQL создать объединенный датасет со значениями из двух таблиц БД, т.е. чтобы у каждого района было известно среднее арифметическое медианной стоимости жилья для районов в его кластере. Условие: только с помощью SQL, не с помощью Python;

```python
def create_new_table(db_name):
    
    ### создадим новую таблицу result в нашей базе данных

    engine_v2 = create_engine(f'sqlite:///{db_name}')

    with engine_v2.begin() as conn:
        sql = """\
            CREATE TABLE result AS
            SELECT *
            FROM neighborhoods
            JOIN cluster_mean_values
            ON neighborhoods.ocean_proximity = cluster_mean_values.ocean_proximity_cluster
            """
        conn.execute(sql)
```

```python
# убедимся, что новая таблица results и отредактированная таблица neighborhoods 
# корректно прочитываются с помощью pandas

def get_new_tables():

    engine = create_engine('sqlite:///test_project_filled.db')

    n_sql = """SELECT id, longitude, latitude, median_house_value, population, ocean_proximity 
               FROM neighborhoods"""
    neighborhoods_df = pd.read_sql(n_sql, engine)

    r_sql = """SELECT id, longitude, latitude, median_house_value, population, ocean_proximity, mean_median_house_value  
               FROM result"""
    result_df = pd.read_sql(r_sql, engine)

    return neighborhoods_df, result_df
```

### TASK 4

- Определить форму распределения **median_house_value** в каждом кластере. 
- Определить выборочные средние **median_house_value** и их отклонения от средних ген. совокупностей. 

Все изобразить графически для каждого кластера. В виде комментария описать ваши выводы о репрезентативности выборок;

```python
def visualize_median_house_value(neighborhoods_filled_df, cluster_mean_values):

    for _, row in cluster_mean_values.iterrows():

        cluster = row.ocean_proximity_cluster
        mean = row.mean_median_house_value

        x = neighborhoods_filled_df[neighborhoods_filled_df["ocean_proximity"] == cluster].median_house_value
        sample_mean = x.mean()
        sample_median = x.median()
        std = round(mean - sample_mean)
        
        print(f"{cluster}. Отклонение выборочного среднего: {std}")
        plt.hist(x, bins=100)

        plt.xlabel('median_house_value')
        plt.ylabel('Count')

        plt.axvline(mean, color='g', linestyle='-', label="mean")
        plt.axvline(sample_mean, color='r', linestyle='--', label="sample_mean")
        plt.axvline(sample_median, color='b', linestyle='--', label="sample_median")
        
        plt.legend()
        plt.show()
```

```python
visualize_median_house_value(neighborhoods_filled_df, cluster_mean_values)
```

Кластер ISLAND в силу своего размера является нерепрезентативной выбокой. Его выборочное среднее сильно отклоняется от среднего генеральной совокупности. Для всех остальных кластеров отклонение незначительно, что говорит о репрезентативности выборок. Кроме ISLAND, все выборки имеют распределение с хвостом справа. INLAND относительно остальных - более узкое и высокое.

### TASK 5

Определить зависимость стоимости жилья от численности населения района в каждом кластере. Описать в виде комментария;

```python
def visualize_pop_vs_value(neighborhoods_filled_df, cluster_mean_values):
    
    for _, row in cluster_mean_values.iterrows():
        
        cluster = row.ocean_proximity_cluster
        data = neighborhoods_filled_df[neighborhoods_filled_df.ocean_proximity == cluster]
        
        print(f"{cluster}. Корреляция: {data.population.corr(data.median_house_value)}")
        plt.scatter(x=data.population, 
                    y=data.median_house_value,
                    alpha=0.25)
        plt.xlabel("population")
        plt.ylabel("median_house_value")
        plt.show()
```

```python
visualize_pop_vs_value(neighborhoods_filled_df, cluster_mean_values)
```

Исходя из графиков и значений корреляции, зависимости стоимости жилья от численности населения района не наблюдается, кроме кластера ISLAND. Однако размер данного кластера слишком мал, чтобы делать выводы.
